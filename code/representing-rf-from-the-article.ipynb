# One-hot encoding

```python
#list of all aminoacids
aminoacids = np.array([
    ["A"], ["C"], ["D"],
    ["E"], ["F"], ["G"],
    ["H"], ["I"], ["K"],
    ["L"], ["M"], ["N"],
    ["P"], ["Q"], ["R"],
    ["S"], ["T"], ["V"],
    ["W"], ["Y"]])
```

```python
#encoder
encoder = OneHotEncoder()
encoder = encoder.fit_transform(aminoacids).toarray()
```

```python
encoder_list = []
for i in encoder:
    encoder_list.append(list(i))
encoder_list
```

```python
aminoacids_list = []
for i in aminoacids:
    aminoacids_list.append(i[0])
aminoacids_list
```

```python
enc = dict(zip(aminoacids_list, encoder))
```

```python
enc
```

## Preprocessing data

```python
def encoding_func(dataset):
  encoded_list = []
  for i in range(len(dataset)):
    seq = dataset['junction_aa'].iloc[i]
    seq_encodded = []
    for letter in seq:
      seq_encodded.append(enc[letter])
    seq_encodded = np.array(seq_encodded)
    encoded_list.append(seq_encodded)
  encoded_list = np.array(encoded_list)
  return encoded_list
```

```python
ACE2_train_enc = encoding_func(ACE2_train)
ACE2_test_enc = encoding_func(ACE2_test)
```

```python
ACE2_train_X = ACE2_train_enc.reshape(ACE2_train_enc.shape[0], (ACE2_train_enc.shape[1]*ACE2_train_enc.shape[2]))
ACE2_test_X = ACE2_test_enc.reshape(ACE2_test_enc.shape[0], (ACE2_test_enc.shape[1]*ACE2_test_enc.shape[2]))
```

```python
ACE2_train_y = ACE2_train['Label']
ACE2_test_y = ACE2_test['Label']
```
